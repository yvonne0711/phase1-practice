```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Basic figure
y = np.linspace(-1, 1, 5) #first array [0,1,2,3,4] from -1 to 5 going up in 1s
x = np.arange(5) #number of values

fig = plt.figure()    #create a figure object
ax = fig.subplots()   #create an axes object in the figure (position, grid rows, grid columns)
line1 = ax.plot(x,y)  #create a line
```

```python
y = np.linspace(-1, 1, 5) #first array [-1, -0.5, 0, 0.5, 1] going from -1 to 1 increments of 5
x = np.arange(5) #number of values array being [0, 1, 2, 3, 4]

fig = plt.figure()    #create a figure object
ax = fig.subplots()   #create an axes object in the figure (position, grid rows, grid columns)
line1 = ax.plot(x,y) 

y2 = np.linspace(-2, 2, 5)
line2 = ax.plot(x, y2)
#ax variable is an axes object so somewhat acts a container for our plots
#y axis auto adjusted from [-1, 1] to [-2, 2] with the new line
```

```python
#6 subplots in 3x2 grid
number_rows = 3
number_columns = 2
fig2 = plt.figure()
ax2 = fig2.subplots(number_rows, number_columns)

#plot subplot row 1 column 2
ax2[0,1].plot(x,y)
```
